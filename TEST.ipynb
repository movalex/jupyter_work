```python
def isprime(n):
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(n**0.5 + 1), 2):
        if n % i == 0:
            return False
    return True


def buildPrimes(n):
    return (i for i in range(2, n + 1) if isprime(i))


def primeFactors(n):
    primes = []
    x = n
    pi = 1
    for i in buildPrimes(int(n**0.5)):
        c = 0
        if x % i == 0:
            while x % i == 0:
                pi = pi * i
                c += 1
                x /= i
            primes.append((str(i)+"**"+str(c)).replace(r"**1",""))
            c = 0
            other = n // pi
            if isprime(other):
                if x % other == 0:
                    while x % other == 0:
                        c += 1
                        x /= other
                    primes.append((str(other)+"**"+str(c)).replace(r"**1",""))
    if not primes:
        return f"({n})"
    return f"({')('.join(primes)})"
```
